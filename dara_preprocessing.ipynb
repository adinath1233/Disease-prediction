```python
import pandas as pd
import numpy as np

#main dataset
dataset_main=pd.read_csv('dataset.csv')
dataset_main.T
```

```python
for col in dataset_main.columns:
    dataset_main[col] = dataset_main[col].str.replace(' ', '')
    dataset_main[col] = dataset_main[col].str.replace('_', ' ')
```

```python
symptoms_list=[]
for i in range(len(dataset_main['Disease'].values)):
    symptoms_list.append([data for data in dataset_main.T[i].values[1:4] if type(data) == str])
```

```python
symptoms_list
```

```python
symptoms_dict={'disease': dataset_main['Disease'].values, 'symptoms':symptoms_list}
health_probelm=pd.DataFrame(symptoms_dict)
health_probelm
```

```python
health_probelm.to_csv('cleaned_dataset.csv', index=False)
```

```python
df = pd.read_csv('cleaned_dataset.csv')

symptoms = df['symptoms'].apply(eval).apply(pd.Series)
symptom_columns = ['symptom{}'.format(i + 1) for i in range(symptoms.shape[1])]
symptoms.columns = symptom_columns

# Concatenate the original dataset with the new symptom columns
new_df = pd.concat([df.drop('symptoms', axis=1), symptoms], axis=1)

# Display the new dataset
print(new_df.head())
```

```python
new_df.to_csv('final_dataset.csv', index=False)
```
